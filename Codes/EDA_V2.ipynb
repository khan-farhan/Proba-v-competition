Loading low resolution images and their corresponding quality maps for a scene

```python
lr_qm_images = load_lr_qm('../Data/train/RED/imgset0530')
```

```python
fig=plt.figure(figsize=(20,10))
fig.suptitle('Low resolution images and their corresponding quality map for scene imgset0530', fontsize=16)
j = 1
for i in range(0,6,2):
    fig.add_subplot(3, 4, j)
    plt.imshow(lr_qm_images[i][0],cmap=None) 
    
    fig.add_subplot(3, 4, j+1)
    plt.imshow(lr_qm_images[i][1],cmap=None)
    
    fig.add_subplot(3, 4, j +2)
    plt.imshow(lr_qm_images[i+1][0],cmap=None) 
    
    fig.add_subplot(3, 4, j+3)
    plt.imshow(lr_qm_images[i+1][1],cmap=None)
    
    j += 4

            
```
